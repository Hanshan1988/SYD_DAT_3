## Exercise Four - Movie Lens Data

```python
import pandas as pd
```

```python
# Read 'u.user'
user_cols = ['age', 'gender', 'occupation', 'user_id']
users = pd.read_table('u.user', sep='|', header=None, names=user_cols)
users.head()
```

```python
# for each occupation in 'users', count the number of occurrences
users.occupation.value_counts()
```

```python
# for each occupation, calculate the mean age
users.groupby(['occupation'])['age'].mean()
```

```python
# for each occupation, calculate the minimum and maximum ages
users.groupby(['occupation'])['age'].agg({'min_age':min,'max_age':max})
```

```python
# for each combination of occupation and gender, calculate the mean age
users.groupby(['occupation','gender'])['age'].mean()
```

```python
# randomly sample a DataFrame
import random
# randomly sampled 10 observations
rows = random.sample(users.index, 10)
users_10 = users.ix[rows]
print users_10
```
